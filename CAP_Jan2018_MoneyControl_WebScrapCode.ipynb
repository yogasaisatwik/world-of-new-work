```python
#required libraries for web scrap
from urllib.request import urlopen  #to get access of webpage
from bs4 import BeautifulSoup       #to get html content
```

```python
#importing libraries
import pandas as pd
import numpy as np
```

## Web-Scrapping

```python
Headers = "Date,Recommendation"
list_recommendation = []
for page in range(200,210):
    url = "https://www.moneycontrol.com/broker-research/markets/cash-"+str(page)+".html".format(page)
    html = urlopen(url)                                  # to get access of webpage
    soup = BeautifulSoup(html,"html.parser")             #  to get html content from webpage
    Title = soup.find_all("div", {"class":"Ohidden"})   # to get all division of recommendations    
    for i in Title:
        try:
            date = i.find("p", {"class":"op_gl12"}).get_text()           #extracting date
            recommendation = i.find("p", {"class":"MT5"}).get_text()     # extracting recommendation
            list_recommendation.append((date,recommendation))
            
        except:
            pass
```

```python
list_recommendation
```

```python
for i in range(len(list_recommendation)):
    a = list_recommendation[i][0].split('|')[1] + ";"
    list_recommendation[i] = (a + list_recommendation[i][1])
```

```python
list_recommendation
```

```python
#cleaning the 
for i in range(len(list_recommendation)):
    a = list_recommendation[i].lower()
    if ":" in a:
        a = a.replace(":",";")
    if "accumulate" in a:
        a = a.replace("accumulate","Buy")
    if "Pick" in a:
        a = a.replace("pick","Buy")
    list_recommendation[i] = a
    
```

```python
z = list_recommendation
z
```

```python
final_list = []
for i in range(len(z)):
    if "target of rs" in z[i]:
        final_list.append(z[i])
```

## Dataframe

```python
#creating dataframea
money_control = pd.DataFrame(columns=["Date","Stock_Name","Action","Target","Recommender"])
```

```python
# functions for merging recommendation in the original dataframe
def appending_data(info_list):
    j = 0
    row_num = money_control.shape[0]+1
    for i in money_control.columns:
        money_control.loc[row_num,i] = info_list[j]
        j = j+1
```

```python
money_control
```

```python
for i in final_list:
    dummy_list = []
    a = i.split(";")
    dummy_list.append(a[0])
    dummy_list.append(a[1][a[1].find(" "):])
    dummy_list.append(a[1][:a[1].find(" ")])
    try:
        dummy_list.append([int(s) for s in a[2].split() if s.isdigit()][0])
    except:
        pass
    try:
        dummy_list.append(a[3])
    except:
        pass
    #print (dummy_list)
    if len(dummy_list)==5:
        appending_data(dummy_list)
```
